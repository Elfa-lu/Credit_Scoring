```python
import pandas as pd
import numpy as np
import math
import time

import warnings
warnings.filterwarnings("ignore")
```

```python
pd_data = pd.read_csv('mx_test_data.csv')
pd_data
```

```python
period = (('10d',10),
          ('30d',30),
          ('90d',90),
          ('180d',180),
          ('360d',360)
          #,('all',2**31 - 1)
         )
```

```python
duration = (('0',0),
       ('30',30),
       ('60',60),
       ('120',120))
```

```python
pd_data['time_1'] = pd.to_datetime(pd_data['time'].apply(lambda x:x[:10]),format="%Y-%m-%d")
pd_data['get_time'] = pd.to_datetime(pd_data['get_time'],format="%Y%m%d")
pd_data['date_diff'] = pd_data[['get_time','time_1']].apply(lambda x:(x[0]-x[1]).days,axis=1)
pd_data['time_hour'] = pd_data['time'].apply(lambda x:x[11:13]).astype(int) # x点
pd_data['time_dayofweek'] = pd_data['time_1'].dt.dayofweek   # 周x
```

```python
id_data = pd.DataFrame(index=pd_data['cert_no'].drop_duplicates().values)
```

### 通话

```python
for prdname, prdtime in period:
    
    # 10/30/90/180/360天内有效通话（时长>=0/30/60/120）总次数/总时长/最大时长
    for durname, durnum in duration:     
        prddata = pd_data[(pd_data['duration']>=durnum) & (pd_data['date_diff']<prdtime)]
        id_data['mx_n_gt_' + durname + '_' + prdname] = prddata[['cert_no','duration']].groupby('cert_no').count()
        id_data['mx_sum_gt_' + durname + '_' + prdname] = prddata[['cert_no','duration']].groupby('cert_no').sum()
        id_data['mx_max_gt_' + durname + '_' + prdname] = prddata[['cert_no','duration']].groupby('cert_no').max()
    
    # 10/30/90/180/360天内通话（通话时间间隔2小时）总次数/总时长/最大时长
    prddata = pd_data[pd_data['date_diff']<prdtime]
    for i in range(0,24,2):
        id_data['mx_n_hour_' + str(i) + '_' + str(i+2) + '_' + prdname] = \
            prddata[['cert_no','duration']][(pd_data['time_hour']>=i) & (pd_data['time_hour']<(i+2))].groupby('cert_no').count()
        id_data['mx_sum_hour_' + str(i) + '_' + str(i+2) + '_' + prdname] = \
            prddata[['cert_no','duration']][(pd_data['time_hour']>=i) & (pd_data['time_hour']<(i+2))].groupby('cert_no').sum()
        id_data['mx_max_hour_' + str(i) + '_' + str(i+2) + '_' + prdname] = \
            prddata[['cert_no','duration']][(pd_data['time_hour']>=i) & (pd_data['time_hour']<(i+2))].groupby('cert_no').max()

    
    # 10/30/90/180/360天内通话（通话时间间隔4小时）总次数/总时长/最大时长
    for i in range(0,24,4):
        id_data['mx_n_hour_' + str(i) + '_' + str(i+4) + '_' + prdname] = \
            prddata[['cert_no','duration']][(pd_data['time_hour']>=i) & (pd_data['time_hour']<(i+4))].groupby('cert_no').count()
        id_data['mx_sum_hour_' + str(i) + '_' + str(i+4) + '_' + prdname] = \
            prddata[['cert_no','duration']][(pd_data['time_hour']>=i) & (pd_data['time_hour']<(i+4))].groupby('cert_no').sum()
        id_data['mx_max_hour_' + str(i) + '_' + str(i+4) + '_' + prdname] = \
            prddata[['cert_no','duration']][(pd_data['time_hour']>=i) & (pd_data['time_hour']<(i+4))].groupby('cert_no').max()
    
    # 10/30/90/180/360天内通话（通话时间间隔8小时）总次数/总时长/最大时长
    id_data['mx_n_hour_0_8_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=0) & (pd_data['time_hour']<8)].groupby('cert_no').count()
    id_data['mx_n_hour_8_16_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=8) & (pd_data['time_hour']<16)].groupby('cert_no').count()
    id_data['mx_n_hour_16_24_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=16) & (pd_data['time_hour']<24)].groupby('cert_no').count()
    id_data['mx_sum_hour_0_8_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=0) & (pd_data['time_hour']<8)].groupby('cert_no').sum()
    id_data['mx_sum_hour_8_16_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=8) & (pd_data['time_hour']<16)].groupby('cert_no').sum()
    id_data['mx_sum_hour_16_24_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=16) & (pd_data['time_hour']<24)].groupby('cert_no').sum()
    id_data['mx_max_hour_0_8_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=0) & (pd_data['time_hour']<8)].groupby('cert_no').max()
    id_data['mx_max_hour_8_16_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=8) & (pd_data['time_hour']<16)].groupby('cert_no').max()
    id_data['mx_max_hour_16_24_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=16) & (pd_data['time_hour']<24)].groupby('cert_no').max()
    
    # 10/30/90/180/360天内通话（通话时间间隔24小时）总次数/总时长/最大时长
    id_data['mx_n_hour_0_24_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=0) & (pd_data['time_hour']<24)].groupby('cert_no').count()
    id_data['mx_sum_hour_0_24_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=0) & (pd_data['time_hour']<24)].groupby('cert_no').sum()
    id_data['mx_max_hour_0_24_' + prdname] = prddata[['cert_no','duration']][(pd_data['time_hour']>=0) & (pd_data['time_hour']<24)].groupby('cert_no').max()
    
    # 10/30/90/180/360天内主叫总次数/总时长/最大时长
    id_data['mx_n_DIAL_' + prdname] = prddata[['cert_no','duration']][pd_data['dial_type'] == 'DIAL'].groupby('cert_no').count()
    id_data['mx_sum_DIAL_' + prdname] = prddata[['cert_no','duration']][pd_data['dial_type'] == 'DIAL'].groupby('cert_no').sum()
    id_data['mx_max_DIAL_' + prdname] = prddata[['cert_no','duration']][pd_data['dial_type'] == 'DIAL'].groupby('cert_no').max()
    
    # 10/30/90/180/360天内被叫总次数/总时长/最大时长
    id_data['mx_n_DIALED_' + prdname] = prddata[['cert_no','duration']][pd_data['dial_type'] == 'DIALED'].groupby('cert_no').count()
    id_data['mx_sum_DIALED_' + prdname] = prddata[['cert_no','duration']][pd_data['dial_type'] == 'DIALED'].groupby('cert_no').sum()
    id_data['mx_max_DIALED_' + prdname] = prddata[['cert_no','duration']][pd_data['dial_type'] == 'DIALED'].groupby('cert_no').max()
       
    # 10/30/90/180/360天内工作日通话总次数/总时长/最大时长
    id_data['mx_n_Workday_' + prdname] = prddata[['cert_no','duration']][pd_data['time_dayofweek'].isin([0,1,2,3,4])].groupby('cert_no').count()
    id_data['mx_sum_Workday_' + prdname] = prddata[['cert_no','duration']][pd_data['time_dayofweek'].isin([0,1,2,3,4])].groupby('cert_no').sum()
    id_data['mx_max_Workday_' + prdname] = prddata[['cert_no','duration']][pd_data['time_dayofweek'].isin([0,1,2,3,4])].groupby('cert_no').max()
    
    # 10/30/90/180/360天内周末通话总次数/总时长/最大时长
    id_data['mx_n_Weekend_' + prdname] = prddata[['cert_no','duration']][pd_data['time_dayofweek'].isin([5,6])].groupby('cert_no').count()
    id_data['mx_sum_Weekend_' + prdname] = prddata[['cert_no','duration']][pd_data['time_dayofweek'].isin([5,6])].groupby('cert_no').sum()
    id_data['mx_max_Weekend_' + prdname] = prddata[['cert_no','duration']][pd_data['time_dayofweek'].isin([5,6])].groupby('cert_no').max()
```

### 联系人

```python
for prdname, prdtime in period:
    
    # 10/30/90/180/360天内联系人数（时长>=0/30/60/120）
    for durname, durnum in duration:
        prddata = pd_data[(pd_data['duration']>=durnum) & (pd_data['date_diff']<prdtime)]
        id_data['mx_n_contact_gt_' + durname + '_' + prdname] = prddata[['cert_no','peer_number']].drop_duplicates().groupby('cert_no').count()
        
    # 10/30/90/180/360天内联系人数（主叫，时长>=0/30/60/120）
    for durname, durnum in duration:
        prddata = pd_data[(pd_data['duration']>=durnum) & (pd_data['date_diff']<prdtime) & (pd_data['dial_type'] == 'DIAL')]
        id_data['mx_n_contact_dial_gt_' + durname + '_' + prdname] = prddata[['cert_no','peer_number']].drop_duplicates().groupby('cert_no').count()
        
    # 10/30/90/180/360天内联系人数（被叫，时长>=0/30/60/120）
    for durname, durnum in duration:
        prddata = pd_data[(pd_data['duration']>=durnum) & (pd_data['date_diff']<prdtime) & (pd_data['dial_type'] == 'DIALED')]
        id_data['mx_n_contact_dialed_gt_' + durname + '_' + prdname] = prddata[['cert_no','peer_number']].drop_duplicates().groupby('cert_no').count()
        
    # 10/30/90/180/360天内联系人数（双向，时长>=0/30/60/120）
    for durname, durnum in duration:
        prddata = pd_data[(pd_data['duration']>=durnum) & (pd_data['date_diff']<prdtime)]
        a = prddata[['cert_no','peer_number','dial_type']].drop_duplicates().groupby(['cert_no','peer_number']).count()
        id_data['mx_n_contact_mutual_gt_' + durname + '_' + prdname] = a[a['dial_type'] == 2].groupby('cert_no').count()
 

    # 10/30/90/180/360天内联系人最大通话数（时长>=0/30/60/120）
    for durname, durnum in duration:
        prddata = pd_data[(pd_data['duration']>=durnum) & (pd_data['date_diff']<prdtime)]
        id_data['mx_max_contact_gt_' + durname + '_' + prdname] = prddata[['cert_no','peer_number','duration']].groupby(['cert_no','peer_number']).count().groupby('cert_no').max()
            #count().sort_values(by=['cert_no','duration']).groupby('cert_no').tail(1)
    
    # 10/30/90/180/360天内联系人最大通话数（主叫，时长>=0/30/60/120）
    for durname, durnum in duration:
        prddata = pd_data[(pd_data['duration']>=durnum) & (pd_data['date_diff']<prdtime) & (pd_data['dial_type'] == 'DIAL')]
        id_data['mx_max_contact_dial_gt_' + durname + '_' + prdname] = prddata[['cert_no','peer_number','duration']].groupby(['cert_no','peer_number']).count().groupby('cert_no').max()
        
    # 10/30/90/180/360天内联系人最大通话数（被叫，时长>=0/30/60/120）
    for durname, durnum in duration:
        prddata = pd_data[(pd_data['duration']>=durnum) & (pd_data['date_diff']<prdtime) & (pd_data['dial_type'] == 'DIALED')]
        id_data['mx_max_contact_dialed_gt_' + durname + '_' + prdname] = prddata[['cert_no','peer_number','duration']].groupby(['cert_no','peer_number']).count().groupby('cert_no').max()

```

```python
# 10/30/90/180/360天内联系人最大通话数（双向，时长>=0/30/60/120）
id_dial = pd.DataFrame(index=pd_data[['cert_no','peer_number']].drop_duplicates().values)
id_dialed = pd.DataFrame(index=pd_data[['cert_no','peer_number']].drop_duplicates().values)

for prdname, prdtime in period:
    for durname, durnum in duration:
        prddata = pd_data[(pd_data['duration']>=durnum) & (pd_data['date_diff']<prdtime)]
        id_dial['dial_num_' + prdname + '_' + durname] = prddata[['cert_no','peer_number','dial_type']][prddata['dial_type']=='DIAL'].groupby(['cert_no','peer_number']).count()
        id_dialed['dialed_num_' + prdname + '_' + durname] = prddata[['cert_no','peer_number','dial_type']][prddata['dial_type']=='DIALED'].groupby(['cert_no','peer_number']).count()

id_dial = id_dial.reset_index().rename(columns={'index':'cert_peer'})
id_dialed = id_dialed.reset_index().rename(columns={'index':'cert_peer'})

id_mutual = pd.merge(id_dial, id_dialed)
id_mutual['cert_no'] = id_mutual['cert_peer'].map(lambda x: x[0])
id_mutual['peer_number'] = id_mutual['cert_peer'].map(lambda x: x[1])

#id_mutual.dropna(axis=0, how='any', inplace=True)
for prdname, prdtime in period:
    for durname, durnum in duration:
        id_mutual['mutual_dial_' + prdname + '_' + durname] = id_mutual['dial_num_' + prdname + '_' + durname] + id_mutual['dialed_num_' + prdname + '_' + durname]
        id_data['mx_max_contact_mutual_gt_'+ durname + '_' + prdname] = id_mutual[['mutual_dial_' + prdname + '_' + durname, 'cert_no']].groupby('cert_no').max()
```

```python
id_data.fillna(0,inplace=True)
id_data = id_data.reset_index().rename(columns={'index':'cert_no'})
id_data.to_csv('mx_character.csv',index=False)
```
